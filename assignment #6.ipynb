```python
# Question.1
# what is boject oriented programming language?
# Answer.1
# Object-oriented programming (OOP) refers to a type of computer programming (software design) 
# in which programmers define the data type of a data structure, and also the types of operations 
# (functions) that can be applied to the data structure.
#In this way, the data structure becomes an object that includes both data and functions. In addition,
#programmers can create relationships between one object and another. For example, objects can inherit
#characteristics from other objects.
```

```python
# Question.2
# list down the benefits of (OOP)?
# Answer.2 
#   There are four advantages of (OOP):
#   1. Modularity For easier troubleshooting
#   2. Reuse of code through inheritance
#   3. Flexibility through polymorphism
#   4. Effective problem solving
```

```python
#class
class Car():
    #instantiation
    def __init__ (self,name,colour,model,brakes):
        #attributes
        self.name=name
        self.colour=colour
        self.model=model
        self.brakes=brakes
        #default attribute
        self.cc='1000 cc'
        
     #methods   
    def updateccnumber(self,newcc):
        self.cc=newcc
        
        
    def updatebrakes(self,newbrakes):
        self.brakes="drum brakes"
        
        
    def updatecolour(self,newcolour):
        self.colour="black colour"
        
#instance #1  
print ("    1st car    ")
people1 = Car('nissan sunny','white',1983,'disc brakes')
print(people1.name)
print(people1.colour)
print(people1.model) 
print(people1.brakes)
print (people1.cc,'old cc')
people1.updateccnumber('1300')
print(people1.cc,'modified cc')
print('\n')
#instance # 2
print("        2nd car")
people2 = Car('Lamboghini','red',2016,'disc brakes')
print(people2.name)
print(people2.colour)
print(people2.model) 
print(people2.brakes,'old brakes')
print (people1.cc)
people1.updatebrakes('drum brakes')
print(people1.brakes,'modified brakes')
print('\n')
print("        3nd car")
#instance 3
people3 = Car('bogadi','blue',2017,'disc brakes')
print(people3.name)
print(people3.colour)
print(people3.model) 
print(people3.brakes)
print (people1.cc)
people3.updatecolour('black colour')
print(people3.colour,'modified colour')
print('\n')
#instance 4 
print ("    4st car    ")
people4 = Car('mercedes','silver',2018,'disc brakes')
print(people4.name)
print(people4.colour)
print(people4.model) 
print(people4.brakes)
print (people4.cc)
print('\n')
#instance 5  
print ("    5st car    ")
people5 = Car('toyota corolla','blue',2019,'drum brakes')
print(people5.name)
print(people5.colour)
print(people5.model) 
print(people5.brakes)
print (people5.cc)
print('\n')
```
